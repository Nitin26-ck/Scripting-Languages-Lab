```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.drop(['PassengerId','Name','Parch','Ticket'],axis=1,inplace=True)
df.head(5)
```

```python
df['Survived']=df['Survived'].map({
    0:'Died',
    1:'Survived'
})
df.head(5)
```

```python
df['Pclass']=df['Pclass'].map({
    1:'Luxuary class',
    2:'Economy class',
    3:'Lower class'
})
df.head(5)
```

```python
df.drop(['Cabin','Fare'],axis=1,inplace=True)
df.head(5)
```

```python
df['Embarked']=df['Embarked'].map({
    'C':'Chersbourg',
    'Q':'Queenstown',
    'S':'Southampton'
})
df.head(5)
```

```python
ax=sns.countplot(x='Pclass',hue='Survived',palette='Set1',data=df)
ax.set(title='Passenger status against Passenger class',xlabel='Pclass',ylabel='Total')
plt.show()
```

```python
ax=sns.countplot(x='Sex',hue='Survived',palette='Set2',data=df)
ax.set(title='survived against sex',xlabel='Sex',ylabel='Total')
plt.show()
```

```python
ax=sns.countplot(x='Embarked',hue='Survived',palette='Set1',data=df)
ax.set(title='survived distribution acc emabarked port ',ylabel='Total')
plt.show()
```

```python
intervals=(0,18,35,50,120)
categories=['Children','Teens','Adults','Old']
df['age_cats']=pd.cut(df.Age,intervals,labels=categories)
```

```python
ax=sns.countplot(x='age_cats',hue='Survived',palette='Set1',data=df)
ax.set(title='survived distribution acc to age categories',xlabel='age_categories',ylabel='total')
plt.show()
```
