```python
df.drop(['User_ID','Product_ID','Stay_In_Current_City_Years'],axis=1,inplace=True)
df.head(5)
```

```python
df['City_Category'] = df['City_Category'].fillna("A")
df.head(5)
```

```python
df['City_Category'] = df['City_Category'].map({
    'A':'Metro cities',
    'B':'Small Towns',
    'C':'Villages'
})
df.head(5)
```

```python
df['Marital_Status'] = df['Marital_Status'].map({
    0:'Unmarried',
    1:'Married'
})
df.head(5)
```

```python
df.rename(columns={'Product_Category_1':'Baseball_Caps','Product_Category_2':'Wine_Tumblers','Product_Category_3':'Pet_Raincoats'},inplace=True)
df.head(5)
```

```python
pd.crosstab(df.Gender,df.Pet_Raincoats)
```

```python
ax=sns.countplot(x='Gender',hue='Pet_Raincoats',palette='Set1',data=df)
ax.set(title='distribution of gender acc to pet raincoats',xlabel='Gender',ylabel='Total')
plt.show()
```

```python
pd.crosstab(df.Gender,df.Wine_Tumblers)
```

```python
ax=sns.countplot(x='Gender',hue='Wine_Tumblers',palette='Set2',data=df)
ax.set(title='distribution of gender acc to wine tumblers',xlabel='Gender',ylabel='Total')
plt.show()
```

```python
ax=sns.countplot(x='Gender',hue='Baseball_Caps',palette='Set1',data=df)
ax.set(title='distribution of gender acc to baseball caps',xlabel='Gender',ylabel='Total')
plt.show()
```
